# Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Read data from the text file
data = np.loadtxt('Laplance.txt')
x = data[:, 0]
y = data[:, 1]
V = data[:, 2]

# Reshape the data to a grid
Nx = len(np.unique(x))
Ny = len(np.unique(y))
X = x.reshape(Nx, Ny)
Y = y.reshape(Nx, Ny)
Z = V.reshape(Nx, Ny)

# Create the figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Customize the plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('V')
ax.set_title('3D Plot of Laplace Equation Solution')

# Add a color bar
fig.colorbar(surf)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt("Lnorm.txt")
i = data1[:,0]
e = data1[:,1]
plt.plot(i, e, 'maroon')
plt.grid(True)
plt.xlabel("Iterations")
plt.ylabel("Residual")
plt.show()


# plot for solution
from scipy.interpolate import griddata

# Load data from file
data = np.loadtxt('Poisson.txt')
x = data[:,0]
y = data[:,1]
f = data[:,2]

# Interpolate f onto a regular grid
xi = np.linspace(x.min(), x.max(), 1000)
yi = np.linspace(y.min(), y.max(), 1000)
xi, yi = np.meshgrid(xi, yi)
fi = griddata((x, y), f, (xi, yi), method='linear')

    # Create plot
plt.contour(xi, yi, fi, 20, linewidths=0.5, colors='k')
plt.contourf(xi, yi, fi, 20, cmap='viridis')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour plot of Poisson eq')
plt.show()
```

# Problem 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from file
data = np.loadtxt('Heat.txt')  # Assuming the file is named 'data.txt'

# Extract spatial points and time
x = np.arange(len(data[0]))  # Spatial points
t = np.arange(len(data))     # Time

# Create meshgrid for contour plot
X, T = np.meshgrid(x, t)

# Create the contour plot
plt.contourf(X, T, data, cmap='hot')

# Add colorbar and labels
plt.colorbar(label='Temperature')
plt.xlabel('Spatial Points')
plt.ylabel('Time')
plt.title('Contour Plot of Temperature')

# Display the plot
plt.show()
```
