## Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt 

# defining names of files
filename = ["legendre_n1.txt","legendre_n2.txt", "legendre_n3.txt", "legendre_n4.txt", "legendre_n5.txt"]

# Adding labels 
labels = ["P_1(x)","P_2(x)", "P_3(x)", "P_4(x)", "P_5(x)"]

for i in range(len(filename)): 
    # loading data from files 
    data = np.loadtxt(filename[i])
    x = data[:,0]
    y = data[:,1]
    
    # Plotting the data
    plt.plot(x,y, label = labels[i])
    
    # Adding labels
    plt.xlabel("x")
    plt.ylabel("P_n(x)")
    plt.grid(True)
    #plt.ylim([-0.5, 1.0])
    plt.xlim([-1.0, 1.0])

# Addding legend and displaying plot
plt.legend()
plt.show()    
    
```

# PROBLEM 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

filename = ["Output.txt", "Output1.txt"]
for i in range(len(filename)):
    # Load data from file
    data = np.loadtxt(filename[i])
    x = data[:,0]
    y = data[:,1]
    f = data[:,2]

    # Interpolate f onto a regular grid
    xi = np.linspace(x.min(), x.max(), 1000)
    yi = np.linspace(y.min(), y.max(), 1000)
    xi, yi = np.meshgrid(xi, yi)
    fi = griddata((x, y), f, (xi, yi), method='linear')

    # Create plot
    plt.contour(xi, yi, fi, 20, linewidths=0.5, colors='k')
    plt.contourf(xi, yi, fi, 20, cmap='viridis')
    plt.colorbar()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Contour plot of ' + filename[i])
    plt.show()
```
