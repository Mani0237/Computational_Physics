# Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt

# defining names of files
filename = ["result_0.200000.txt", "result_0.500000.txt", "result_0.700000.txt", "result_1.200000.txt", "result_1.700000.txt"]

# Adding labels 
labels = ["N_0 = 0.2", "N_0 = 0.5", "N_0 = 0.7", "N_0 = 1.2", "N_0 = 1.7"]

for i in range(len(filename)): 
    # loading data from files 
    data = np.loadtxt(filename[i])
    x = data[:,0]
    y = data[:,1]
    
    # Plotting the data
    plt.plot(x,y, label = labels[i])
    
    # Adding labels
    plt.xlabel("Time")
    plt.ylabel("Population(N)")
    plt.grid(True)
    plt.xlim([0.0, 15.0])
    

# Addding legend and displaying plot
plt.legend()
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the text file
data = np.loadtxt('output_problem2.txt')

# Separate the columns into time, x, and y arrays
time = data[:, 0]
x = data[:, 1]
y = data[:, 2]

# Reshape the data into 22 trajectories
num_trajectories = 22
num_rows = 999
time = time.reshape((num_trajectories, num_rows))
x = x.reshape((num_trajectories, num_rows))
y = y.reshape((num_trajectories, num_rows))

# Plot the trajectories
for i in range(num_trajectories):
    plt.plot(x[i], y[i], label=f'Trajectory {i+1}')

plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Trajectories')
#plt.legend()
plt.grid(True)
plt.show()
```

# Problem 5

```python
import numpy as np
import matplotlib.pyplot as plt

# loading data from the file
data = np.loadtxt("problem5_result.txt")
t = data[:, 0]
x = data[:, 1]
y = data[:, 2]
z = data[:, 3]

plt.plot(t, x)
plt.xlim([0, 100])
plt.title("y(t) vs t")
plt.xlabel("Time(in sec)")
plt.ylabel("y(t)")
plt.show()

plt.plot(t, y)
plt.xlim([0, 100])
plt.title("y(t) vs t")
plt.xlabel("Time(in sec)")
plt.ylabel("y(t)")
plt.show()

plt.plot(t, z)
plt.xlim([0, 100])
plt.title("z(t) vs t")
plt.xlabel("Time(in sec)")
plt.ylabel("z(t)")
plt.show()

plt.plot(x, z, 'maroon', linewidth = 0.5 )
plt.xlim([-20, 20])
plt.title("z(t) vs x")
plt.xlabel("x")
plt.ylabel("z(t)")
plt.show()
```
